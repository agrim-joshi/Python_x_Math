### Notice the difference in outputs!

````python
x = 4.6
y = 2.3
display(Math(f"{x} \\times {y} = {x*y}"))
````

````python
x = 4.6
y = 2.3
display(Math("%g \\times %g = %g"%(x,y,x*y)))
````

### Explanation:

This is a **floating-point precision** issue in Python.

Your code:

```python id="jhrj4o"
x = 4.6
y = 2.3

display(Math(f"{x} \\times {y} = {x*y}"))
```

shows:

```text id="egie5y"
10.579999999999998
```

because computers do **not store most decimal numbers exactly** in binary.

---

#### What's happening internally

Python stores numbers in binary (base 2), not decimal (base 10).

Values like:

* `4.6`
* `2.3`

cannot be represented perfectly in binary floating-point format.

So internally Python stores approximations like:

```text id="4w42xv"
4.5999999999999996
2.2999999999999998
```

When multiplied:

```text id="1p0q6t"
10.579999999999998
```

That’s the actual stored floating-point result.

---

#### Why `%g` shows `10.58`

This line:

```python id="2l1wws"
"%g \\times %g = %g"%(x,y,x*y)
```

uses `%g` formatting.

`%g` automatically:

* rounds numbers
* removes unnecessary digits
* shows a cleaner representation

So Python formats:

```text id="0fegk0"
10.579999999999998
```

as:

```text id="vtfqj5"
10.58
```

---

#### Key difference

| Method     | Behavior                        |
| ---------- | ------------------------------- |
| `f"{x*y}"` | Shows full floating-point value |
| `%g`       | Formats/rounds nicely           |

---

#### Better modern solution

Use formatting inside f-strings:

```python id="5c5tnp"
display(Math(f"{x} \\times {y} = {x*y:.2f}"))
```

`.2f` means:

> show 2 decimal places

Output:

4.6\times2.3=10.58

---

#### Useful formatting examples

| Format | Meaning              | Example  |
| ------ | -------------------- | -------- |
| `.2f`  | 2 decimal places     | `10.58`  |
| `.3f`  | 3 decimal places     | `10.580` |
| `.1f`  | 1 decimal place      | `10.6`   |
| `.0f`  | no decimals          | `11`     |
| `.3g`  | 3 significant digits | `10.6`   |

---

#### Important concept

This is not a bug in Python specifically.

It happens in:

* Python
* JavaScript
* C/C++
* Java
* MATLAB
* most programming languages

because they use IEEE floating-point arithmetic.


````python
display(Math(f"{x} \\times {y} = {x*y:.2f}"))
````

## Absolute Value in Mathematics

Absolute value represents the distance of a number from zero on the number line. Since distance is always non-negative, the absolute value of a number is always positive or zero.

The absolute value of a real number x is denoted by:

∣x∣

Examples

∣5∣=5

∣−5∣=5

∣0∣=0


````python
a = -4
b = abs(a)  #  abs() function returns the absolute value
a,b
````

## Displaying absolute values in LaTeX

````python
x = -4.3

display(Math(f"|{{{x}}}| = {abs(x)}"))


# You can also use `%g` placeholder instead of `f-strings` to avoid multiple `{}`

display(Math("|%g| = %g"%(x,abs(x))))  
````

## Displaying text in LaTeX

`\\text{your text here}`

````python
base = 323
for i in range(1,10):
    display(Math(f"{base}^{{{i}}} \\text{{ is }}  {base**i}")) #%g can also be used as a placeholder here
````

## Common Python Format Specifiers

| Placeholder | Meaning | Example |
|------------|---------|---------|
| `%s` | String | `"Hello %s" % "World"` → `Hello World` |
| `%d` | Integer (decimal) | `"%d" % 42` → `42` |
| `%i` | Integer | `"%i" % 42` → `42` |
| `%f` | Floating-point number (6 decimal places by default) | `"%f" % 3.14` → `3.140000` |
| `%.2f` | Float with 2 decimal places | `"%.2f" % 3.14159` → `3.14` |
| `%g` | General format (removes unnecessary zeros) | `"%g" % 10.5800` → `10.58` |
| `%e` | Scientific notation (lowercase e) | `"%e" % 1234` → `1.234000e+03` |
| `%E` | Scientific notation (uppercase E) | `"%E" % 1234` → `1.234000E+03` |
| `%x` | Hexadecimal (lowercase) | `"%x" % 255` → `ff` |
| `%X` | Hexadecimal (uppercase) | `"%X" % 255` → `FF` |
| `%o` | Octal | `"%o" % 8` → `10` |
| `%%` | Literal percent sign | `"100%%"` → `100%` |
