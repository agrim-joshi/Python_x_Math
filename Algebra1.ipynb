# Solving for x

an equation written like:
$$ 2x^2 + 4x - 9 = 0 $$ 

is called an equation in its homogenous form.

````python
from IPython.display import display, Math
import sympy as sym
````

````python
x = sym.symbols('x')

expr = 2*x + 4 - 9  # the equation is in homogenous form implicitly

print(sym.solve(expr)) # solves for x

display(Math("\\text{The solution to } %s \\text{ is } %s"%(sym.latex(expr),sym.latex(sym.solve(expr)))))
````

## How does python know I want to solve for x??

If no symbols are indicated to be of interest and the equation is
univariate, a list of values is returned; otherwise, the keys in
a dictionary will indicate which (of all the variables used in
the expression(s)) variables and solutions were found:

    >>> solve(x**2 - 4)
    [-2, 2]
    >>> solve((x - a)*(y - b))
    [{a: x}, {b: y}]
    >>> solve([x - 3, y - 1])
    {x: 3, y: 1}
    >>> solve([x - 3, y**2 - 1])
    [{x: 3, y: -1}, {x: 3, y: 1}]



### Displaying an Equation and Its Solution

The following code renders a SymPy expression and its solution as a formatted mathematical statement in a Jupyter notebook:

```python
display(
    Math(
        "\\text{The solution to } %s \\text{ is } %s"
        % (sym.latex(expr), sym.latex(sym.solve(expr)))
    )
)
```

**How it works:**

- `sym.solve(expr)` computes the solution(s) of the expression.
- `sym.latex(expr)` converts the expression to LaTeX.
- `sym.latex(sym.solve(expr))` converts the solution(s) to LaTeX.
- `Math(...)` creates a LaTeX math object.
- `display(...)` renders the result as formatted mathematics in the notebook output.

For example, if `expr = x**2 - 4`, the output will appear as:

$$
\text{The solution to } x^2 - 4 \text{ is } [-2, 2]
$$
